```python
# List of City data 
cities_weather = []
```

```python
def get_response(city):
    response = requests.get(
        f"https://api.openweathermap.org/data/2.5/forecast?q={city}&units=imperial&APPID={api_key}"
    )
    return response
```

```python
def convert_datetime(timestamp):
    date_time = datetime.datetime.fromtimestamp(timestamp)
    date_time = f"{date_time:%Y-%m-%d %H:%M:%S}"
    _date = str(date_time).split()[0]
    _time = str(date_time).split()[1]

    return _date, _time
```

```python
def get_data(newlist, cities):
    for city in cities:
        city_data = get_response(city).json()

        # City Details
        city_details = city_data['city']
        lat = city_details['coord']['lat']
        lon = city_details['coord']['lon']
        population = city_details['population']
        sunrise = city_details['sunrise']
        sunset = city_details['sunset']

        # Conveting timestamp
        sunset_time = convert_datetime(sunset)[1]
        sunrise_time = convert_datetime(sunrise)[1]
        
        city_weather = city_data['list']
        for dt in range(0,len(city_weather)):
            dt_data = city_weather[dt]

            # Getting Date and Time
            date_time = dt_data['dt']
            _date = convert_datetime(date_time)[0]
            _time = convert_datetime(date_time)[1]

            # Getting Data
            weather_data = {
            'date' : _date,
            'time' : _time,
            'sunrise_time' : sunrise_time,
            'sunset_time' : sunset_time,
            'city' : city,
            'lat' : lat,
            'lon' : lon,
            'population' : population,
            'weather' : dt_data['weather'][0]['main'],
            'temp' : dt_data['main']['temp'],
            'temp_feels_like' : dt_data['main']['feels_like'],
            'pressure' : dt_data['main']['pressure'],
            'humidity' : dt_data['main']['humidity'],
            'wind' : dt_data['wind']['speed']
            }
            newlist.append(weather_data)
    
    return newlist
```

```python
get_data(cities_weather, cities)
```

```python
# Dataframe - run at the end of the week
weather_data = pd.DataFrame(cities_weather, columns=['date', 'time','sunrise_time', 'sunset_time', 'city','lat', 'lon', 'population', 'weather', 'temp', 'temp_feels_like', 'pressure', 'humidity', 'wind'])
weather_data
```
